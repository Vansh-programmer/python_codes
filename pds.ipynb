Looping Through Strings

```python
fruit = 'apple'
index = 0
while index < len(fruit):
    print(index , fruit[index])
    index+=1
```

```python
fruit = 'apple'
for letter in fruit:
    print(letter)
```

```python
fruit = 'apple'
count = 0
for letter in fruit:
    if letter == 'p':
        count+=1
print(count)
```
