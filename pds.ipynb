```python
fruit = 'apple'
count = 0
for letter in fruit:
    if letter == 'p':
        count+=1
print(count)
```

6.2 Manipulating Strings

```python
word = input('Enter a word:')
if word == 'banana':
    print('all right ' , word)

if word <'banana':
    print('word ' , word , ' is smaller')

if word > 'banana':
    print('word ', word , ' is larger')

else:
    print('Ok')
```
